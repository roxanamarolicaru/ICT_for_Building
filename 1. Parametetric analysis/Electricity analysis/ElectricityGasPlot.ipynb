```python
import matplotlib.pyplot as plt
import os 
import pandas as pd
import numpy as np
```

```python
SG_SH0_D0=pd.read_csv("./dati/SG_SH0_DIMMER0.csv",sep=';')
SG_SH0_D1=pd.read_csv("./dati/SG_SH0_DIMMER1.csv",sep=';')
SG_SH1_D1=pd.read_csv("./dati/SG_SH1_DIMMER1.csv",sep=';')
TG_SH1_D1=pd.read_csv("./dati/TG_SH1_DIMMER1.csv",sep=';')
shading = ["Off", "On"]
glazing =["Single" , "Double", "Triple"]
colors =['red', 'orange', 'green', 'blue']
measures=['Electricty consumption [Wh/m2]','Gas consumption [Wh/m2]']
monthsList = ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "Dicember"]

elListSG_SH0_D0 = list(SG_SH0_D0['Electricity'])[1:14]
elListSG_SH0_D1 = list(SG_SH0_D1['Electricity'])[1:14]
elListSG_SH1_D1 = list(SG_SH1_D1['Electricity'])[1:14]
elListTG_SH1_D1 = list(TG_SH1_D1['Electricity'])[1:14]
gasListSG_SH0_D0 = list(SG_SH0_D0['Gas'])[1:14]
gasListSG_SH0_D1 = list(SG_SH0_D1['Gas'])[1:14]
gasListSG_SH1_D1 = list(SG_SH1_D1['Gas'])[1:14]
gasListTG_SH1_D1 = list(TG_SH1_D1['Gas'])[1:14]

el_SG_SH0_D0=[]
el_SG_SH0_D1=[]
el_SG_SH1_D1=[]
el_TG_SH1_D1=[]
gas_SG_SH0_D0 =[]
gas_SG_SH0_D1 =[]
gas_SG_SH1_D1 =[]
gas_TG_SH1_D1 =[]

for i in elListSG_SH0_D0:
    i= i.replace(",", ".")
    i=float(i)
    el_SG_SH0_D0.append(i)

for i in elListSG_SH0_D1:
    i=i.replace(",", ".")
    i=float(i)
    el_SG_SH0_D1.append(i)

for i in elListSG_SH1_D1:
    i=i.replace(",", ".")
    i=float(i)
    el_SG_SH1_D1.append(i)

for i in elListTG_SH1_D1:
    i=i.replace(",", ".")
    i=float(i)
    el_TG_SH1_D1.append(i)

for i in gasListSG_SH0_D0:
    i= i.replace(",", ".")
    i=float(i)
    gas_SG_SH0_D0.append(i)

for i in gasListSG_SH0_D1:
    i=i.replace(",", ".")
    i=float(i)
    gas_SG_SH0_D1.append(i)

for i in gasListSG_SH1_D1:
    i=i.replace(",", ".")
    i=float(i)
    gas_SG_SH1_D1.append(i)

for i in gasListTG_SH1_D1:
    i=i.replace(",", ".")
    i=float(i)
    gas_TG_SH1_D1.append(i)
```

```python
#Electricty consumption during the year (Single glazing)
plt.plot(monthsList, el_SG_SH0_D0, 'red', label = "Electricity without light dimmering")
plt.plot(monthsList, el_SG_SH0_D1, 'orange', label = "Electricity with light dimmering")
plt.title('Electricity consumption for single glazing windows, without shading control', fontsize=10)
plt.xlabel('Months 2002', fontsize=10)
plt.ylabel('[Wh/m2]', fontsize=10)
plt.legend(loc = 'upper right')
plt.xticks(rotation = 45)
plt.grid()
plt.show()
```

```python
#Electricity consumption during the year for single and triple glazing
plt.plot(monthsList, el_SG_SH1_D1, 'red', label = "Single glazing")
plt.plot(monthsList, el_TG_SH1_D1, 'green', label = "Triple glazing")
plt.title('Electricity consumption for single and triple glazing windows, with shading control', fontsize=10)
plt.xlabel('Months 2002', fontsize=10)
plt.ylabel('[Wh/m2]', fontsize=10)
plt.legend(loc = 'upper right')
plt.xticks(rotation = 45)
plt.grid()
plt.show()
```

```python
# Gas consumption during the year for single and triple glazing
plt.plot(monthsList, gas_SG_SH1_D1, 'red', label = "Single glazing")
plt.plot(monthsList, gas_TG_SH1_D1, 'green', label = "Triple glazing")
plt.title('Gas consumption for single and triple glazing windows', fontsize=10)
plt.xlabel('Months', fontsize=10)
plt.ylabel('[Wh/m2]', fontsize=10)
plt.xticks(rotation = 45)
plt.legend(loc = 'upper right')
plt.grid()
plt.show()
```
